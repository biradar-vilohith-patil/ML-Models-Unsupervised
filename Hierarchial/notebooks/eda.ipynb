```python
import pandas as pd
import numpy as np

df_raw = pd.read_csv('../data/IPL Player Stat.csv')

df_clean = df_raw.copy()
```

```python
df_clean.columns
```

```python

df_clean['batting_avg'] = pd.to_numeric(df_clean['batting_avg'], errors='coerce')
df_clean['batting_strike_rate'] = pd.to_numeric(df_clean['batting_strike_rate'], errors='coerce')
df_clean['boundaries_percent'] = pd.to_numeric(df_clean['boundaries_percent'], errors='coerce')
df_clean['balls_faced'] = pd.to_numeric(df_clean['balls_faced'], errors='coerce')

df_clean = df_clean.dropna(subset=['batting_avg', 'batting_strike_rate', 'boundaries_percent', 'balls_faced'])

df_clean = df_clean[df_clean['balls_faced'] >= 100]

features = ['batting_avg', 'batting_strike_rate', 'boundaries_percent']

```

```python
for col in features:
    Q1 = df_clean[col].quantile(0.05)
    Q3 = df_clean[col].quantile(0.95)
    IQR = Q3 - Q1
    df_clean = df_clean[(df_clean[col] >= Q1 - 1.5 * IQR) & (df_clean[col] <= Q3 + 1.5 * IQR)]

df_final = df_clean[features]

df_final.to_csv('../data/cleaned_ipl_batters.csv', index=False)
```
