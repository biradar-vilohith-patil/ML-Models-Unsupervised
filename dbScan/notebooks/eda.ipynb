```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
df_raw = pd.read_csv('../data/instagram_reach.csv', encoding='latin1')
df_raw.columns
df_raw= df_raw.rename(columns={'Follows':'Followers'})
```

```python
df_clean = df_raw[['Followers', 'Impressions', 'Likes', 'Comments']].copy()
df_clean = df_clean.dropna()
df_clean = df_clean[df_clean['Followers'] > 50]
df_clean = df_clean[df_clean['Impressions'] > 10]
```

```python
df_clean['reach_rate'] = df_clean['Impressions'] / df_clean['Followers']
df_clean['engagement_rate'] = df_clean['Likes'] / df_clean['Impressions']
df_clean['interaction_rate'] = df_clean['Comments'] / df_clean['Likes']

df_clean = df_clean.replace([np.inf, -np.inf], np.nan).dropna()

q1_rr = df_clean['reach_rate'].quantile(0.01)
q9_rr = df_clean['reach_rate'].quantile(0.99)
df_clean = df_clean[(df_clean['reach_rate'] >= q1_rr) & (df_clean['reach_rate'] <= q9_rr)]

df_clean.to_csv('../data/cleaned_creator_metrics.csv', index=False)
```
