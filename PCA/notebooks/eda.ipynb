```python
import pandas as pd
import numpy as np

# Dataset: https://www.kaggle.com/datasets/maharshipandya/-spotify-tracks-dataset
df_raw = pd.read_csv('../data/dataset.csv')

# We only keep the features that map to human emotion/energy
features = ['valence', 'energy', 'acousticness', 'danceability']

df_clean = df_raw[features].copy()
df_clean = df_clean.dropna()

# Remove extreme statistical outliers to make the PCA space clean
for col in features:
    Q1 = df_clean[col].quantile(0.01)
    Q3 = df_clean[col].quantile(0.99)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df_clean = df_clean[(df_clean[col] >= lower_bound) & (df_clean[col] <= upper_bound)]

df_clean.to_csv('../data/cleaned_spotify_vibes.csv', index=False)
```
