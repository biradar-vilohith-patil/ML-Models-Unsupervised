```python
import pandas as pd
import numpy as np

np.random.seed(42)
n_profiles = 2000

fs_frontend = np.random.normal(80, 15, 500)
fs_backend = np.random.normal(75, 15, 500)
fs_data = np.random.normal(30, 15, 500)
fs_cloud = np.random.normal(40, 15, 500)
fs_systems = np.random.normal(20, 10, 500)

ds_frontend = np.random.normal(30, 15, 500)
ds_backend = np.random.normal(60, 15, 500)
ds_data = np.random.normal(85, 10, 500)
ds_cloud = np.random.normal(50, 15, 500)
ds_systems = np.random.normal(30, 15, 500)

do_frontend = np.random.normal(20, 15, 500)
do_backend = np.random.normal(65, 15, 500)
do_data = np.random.normal(40, 15, 500)
do_cloud = np.random.normal(85, 10, 500)
do_systems = np.random.normal(70, 15, 500)

sys_frontend = np.random.normal(10, 10, 500)
sys_backend = np.random.normal(50, 15, 500)
sys_data = np.random.normal(30, 15, 500)
sys_cloud = np.random.normal(40, 15, 500)
sys_systems = np.random.normal(90, 10, 500)

frontend = np.concatenate([fs_frontend, ds_frontend, do_frontend, sys_frontend])
backend = np.concatenate([fs_backend, ds_backend, do_backend, sys_backend])
data = np.concatenate([fs_data, ds_data, do_data, sys_data])
cloud = np.concatenate([fs_cloud, ds_cloud, do_cloud, sys_cloud])
systems = np.concatenate([fs_systems, ds_systems, do_systems, sys_systems])

df_raw = pd.DataFrame({
    'frontend_score': frontend,
    'backend_score': backend,
    'data_score': data,
    'cloud_score': cloud,
    'systems_score': systems
})

df_raw = df_raw.clip(0, 100)
df_clean = df_raw.dropna()

df_clean.to_csv('../data/developer_profiles.csv', index=False)
```
