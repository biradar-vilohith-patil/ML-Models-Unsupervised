```python
import pandas as pd
import numpy as np

df_raw = pd.read_csv('../data/Sleep_health_and_lifestyle_dataset.csv')

df_clean = df_raw[['Sleep Duration', 'Quality of Sleep', 'Physical Activity Level', 'Stress Level', 'Heart Rate']].copy()
df_clean = df_clean.dropna()

for col in df_clean.columns:
    Q1 = df_clean[col].quantile(0.01)
    Q3 = df_clean[col].quantile(0.99)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df_clean = df_clean[(df_clean[col] >= lower_bound) & (df_clean[col] <= upper_bound)]

df_clean.to_csv('../data/cleaned_sleep_metrics.csv', index=False)
```
